# Problem 1 ( 7 pts.)

An ecologist uses a *quadrant sampling technique* to estimate the density of plant population.  A quadrant is a sample plot of a specific size used for the study of a population or a community. Although a quadrant can have many different sizes and shapes, our ecologist uses a simple 100cm x 100cm frame with 20 cm divisions. An example of a quadrant sample with the location of an origin and coordinates of a single flower (flower number 10) is shown below.

During a recent expedition, the ecologist randomly positioned her quadrant frame on the ground and recorded the location of eight flowers within the quadrant. She used the bottom left corner of the quadrant as the origin (*x=0*, *y=0*). The list of lists **flowers** below contains the flower label(1, 2, ...8)  and the x and y coordinates of the flowers within the quadrant. 

```python
flowers =[[1, 90.9,71.8], [2, 9.7, 45.4], [3, 86.7, 22.8], [4, 6.0, 62.6],[5, 82.3, 15.6],[6, 62.6, 48.5],[7, 38.4, 12.4],[8, 93.0, 81.4]]
```

### (A) Find the flower closest to the origin of the quadrant

*Find the label of the flower closest to the bottom left corner (origin).  What is the distance between the origin and the closest flower? Hint: You will use these distances again below, so it might be useful to make a new array containing flower labels and their distances.*

If you cannot remember the formula for calculating the Euclidean distance between two points, here it is:

$d = \sqrt{(x_1 - x_2)^2 +(y_1 - y_2)^2}$

Your code should produce a neat and informative output. Assume that the user of your code has only minimal familiarity with Python: she is only able to run the code and read a printout. The user is not comfortable manipulating or changing your code. Sample printout produced by the code for a different experiment is shown below. Try to keep your output in a similar format.

It was convenient for the ecologist to keep her data as a list of lists. You can choose to either work with the list of lists or convert it into any other variable types studied in Classes 1-4 (list, numpy arrays, etc). However, you are not allowed to re-type the sequence or any of its parts. All type conversions should be done algorithmically (and can be accomplished using the numpy method **astype**, which you can look up)..

```python
import numpy as np

flowersNP = np.array(flowers)

xValues = flowersNP[:,1]
yValues = flowersNP[:,2]

#distances from the origin is calculated
distancesFromOrigin = np.sqrt(xValues ** 2 + yValues ** 2)

#find the minimum value and return the index number
closestToOrigin = np.argmin(distancesFromOrigin)

#find the flower number from the closest to origin
flowerNumber = int(flowersNP[closestToOrigin, 0])

#print the result
print("The flower closest to the origin of the quadrant is flower #:" ,flowerNumber)
```

### (B) Print flower labels for all flowers located farther than 50 cm from the origin of the quadrant

*Print out the labels for the flowers that are more than 50 cm away from the bottom left corner.*

```python
#find boolean values of flowers located farther than 50 cm from the origin of the quadrant
flowersGreaterThan50Boolean = distancesFromOrigin > 50

#numpy array of values located farther than 50 cm from the origin of the quadrant
flowersGreaterThan50 = flowersNP[flowersGreaterThan50Boolean]

#first column of the numpy array
firstColumn = flowersGreaterThan50[:,0]

#values are converted to an int
flowerNumbersINT = firstColumn.astype(int)

#values are converted to an str
flowerNumbersSTR = flowerNumbersINT.astype(str)

flowerNumbers = " ".join(flowerNumbersSTR)

#have it print as just values as int
print("Flowers located farther than 50 cm from the origin of the quadrant: ", flowerNumbers)
```
