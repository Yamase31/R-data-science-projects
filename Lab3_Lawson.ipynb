## (B) Calculate proportionality constant for each volume

The next step is to determine the value of k for each of the beakers of water.  We will look back at our = equation for temperature and then solve for k.

Your goal is to use Python to calculate *k* at different times for each beaker and then find the average *k* value for each beaker. Provide a clear printout of the proportionality constant for each temperature.

**Note**: Be careful to exclude the initial measurement (time zero, temperature 100) from your calculations. Otherwise you will have a division by zero error

```python
#initial measurement is excluded and numpy array is created
t100 = tempC_100ml[1:]
t100np = np.array(t100)

#initial measurement is excluded and numpy array is created
t300 = tempC_300ml[1:]
t300np = np.array(t300)

#initial measurement is excluded and numpy array is created
t800 = tempC_800ml[1:]
t800np = np.array(t800)

#initial measurement is excluded and numpy array is created
time_minNP = np.array(time_min[1:])

#math calculation and average is computed
k100 = np.log((t100np - 23)/(100 - 23))/time_minNP
k100avg = np.average(k100)

#math calculation and average is computed
k300 = np.log((t300np - 23)/(100 - 23))/time_minNP
k300avg = np.average(k300)

#math calculation and average is computed
k800 = np.log((t800np - 23)/(100 - 23))/time_minNP
k800avg = np.average(k800)

#results are printed
print("The average K value of 100ml:", k100avg)
print("The average K value of 300ml:", k300avg)
print("The average K value of 800ml:", k800avg)
```

## (C) Plot experimental and theoretical graphs for each volume on a separate graph. 

You can now use the proportionality constant for your data to make a theoretical graph of the temperature for each volume together with its corresponding experimental data. You can use the formula in the begining of this problem to calculate theoretical values using the mean values of the constant that you calculated for each beaker. Your graphs should be similar to one below with the theoretical values plotted as a line with plt.plot(), and the experimental values plotted as dots with plt.scatter().
```python
#figure size is changed to a 10" x 10"
plt.figure(figsize=(10,10))

#new calculations
temp_100ml_theoretical = 23 + (100 - 23)*np.exp(k100avg*time_minNP) 
temp_300ml_theoretical = 23 + (100 - 23)*np.exp(k300avg*time_minNP)
temp_800ml_theoretical = 23 + (100 - 23)*np.exp(k800avg*time_minNP)

#scatterplot is created with time on the x axis
plt.plot(time_minNP, temp_100ml_theoretical, label="100ml theoretical")
plt.plot(time_minNP, temp_300ml_theoretical, label="300ml theoretical")
plt.plot(time_minNP, temp_800ml_theoretical, label="800ml theoretical")
plt.scatter(time_min, tempC_100ml, label="100ml experimental")
plt.scatter(time_min, tempC_300ml, label="300ml experimental")
plt.scatter(time_min, tempC_800ml, label="800ml experimental")

#title is added
plt.title("Newton's Law of Cooling")

#x and y labels are added
plt.ylabel("Temperature (C)")
plt.xlabel("Time (min)")

#legend
plt.legend(loc="best")

#scatterplot is displayed
plt.show()
```
