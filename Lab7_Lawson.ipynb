## (A) Calculate total number of deaths in each year

## (C) Plotting an ECDF

You probably observed that the shape of the distribution depends on the number of bins in the histogram. To avoid such inconsistency, please plot an ECDF for the total number of deaths. Use <font color='red'> red stars "*" <font color = 'black'> 
as the markers of your data points. You will want to copy in your ECDF function that we have defined in previous classes.

```python
def ecdf(data):
    """Compute ECDF for a one-dimensional array of measurements."""
    # Number of data points: n
    n = len(data)
    # x-data for the ECDF: x
    x = np.sort(data)
    # y-data for the ECDF: y
    y = np.arange(1, n+1) / n
    return x, y

#figure size is set
plt.figure(figsize = (10, 10))

#ecdf function is called on the total deaths data
x, y = ecdf(df.total_deaths)

#ecdf of the total deaths data is plotted
plt.plot(x, y, label="total deaths", color = "red", marker = "*", linestyle = "none")

#legend is created
plt.legend(loc="best")

#title is added
plt.title("ECDFs of Total Deaths", fontsize = 24)

#x and y label are created
plt.xlabel("Total Deaths", fontsize = 14)
plt.ylabel("ECDF", fontsize = 14)

#plot is shown
plt.show()
```

## (D) Calculate the mean number of deaths and use it to generate a Poisson process

Since this dataset is one of the earliest examples of an application of the Poisson distribution, we will fit the data with a Poisson distribution. We need only one parameter to generate a Poisson process: the mean of the distribution. 
Calculate the mean number of total deaths per year in your dataset and print it out:

```python
#mean is calculated
mean = df.total_deaths.mean()

#mean is printed
print("The mean number of total deaths per year:", mean)
```

Use the appropriate normal distribution to generate 10000 random data points to simulate the CDFs for body weight and heart weight in cats (plot a single combined CDF for the male and female cats). ## (D) Linear regression of heart and weight for male cats

Is there a relationship between a body weight and heart weight? Conduct a linear regression of two variables and plot the line of best fit together with a scatterplot of data. Make sure that your graph is clear and easy to follow. You should be able to modify code from the lectures that will work well here. **Remember, we are only interested here in male cats.**

Per each additional kilogram of weight, how much heart mass does the cat gain/lose? Provide an informative printout to asnswer this question in plain English.

```python
#figure size is set
plt.figure(figsize = (10, 10))

#plot the body weight versus heart weight
plt.plot(cats[cats.Sex == "M"].Bwt, cats[cats.Sex == "M"].Hwt, marker='.', linestyle='none')

#perform a linear regression using np.polyfit(): a, b
a, b = np.polyfit(cats[cats.Sex == "M"].Bwt, cats[cats.Sex == "M"].Hwt, 1)

#variables are calculated
x = np.array([1,5])
y = a * x + b

#linear regression line is added to plot
plt.plot(x, y)

#title is added
plt.title("Male Body Weight (kg) vs. Male Heart Weight (g)", fontsize = 24)

#x and y label are created
plt.xlabel("Male Body Weight (kg)", fontsize = 14)
plt.ylabel("Male Heart Weight (g)", fontsize = 14)

#print statement with the slope
print('For each additional kg of body weight, the heart weight of the male cat increases', a, 'g')
print("\n")

#show the plot
plt.show()
```

## (D) Compare linear regression of heart and weight for male and female cats

Now let's compare heart anatomy between male and female cats. Make a scatterplot of heart weight vs. body weight for male and female cats. Use different colors and line markers for males an females. Add regression lines for males and females in appropriate colors. With every additional kilo of body weight, would males or females add more heart mass? How much more? Provide informative printout.

```python
#figure size is set
plt.figure(figsize = (10,10))

#plot with data from males and females is made
plt.plot(cats[cats.Sex == "M"].Bwt, cats[cats.Sex == "M"].Hwt, label="heart weight vs. body weight for male cats", color = "blue", marker='*', linestyle='none')
plt.plot(cats[cats.Sex == "F"].Bwt, cats[cats.Sex == "F"].Hwt, label="heart weight vs. body weight for female cats", color = "orange", marker='.', linestyle='none')

#perform a linear regression for males and females using np.polyfit(): a, b
aMale, bMale = np.polyfit(cats[cats.Sex == "M"].Bwt, cats[cats.Sex == "M"].Hwt, 1)
aFemale, bFemale = np.polyfit(cats[cats.Sex == "F"].Bwt, cats[cats.Sex == "F"].Hwt, 1)

#variables are calculated for male and female
xMale = np.array([1,5])
yMale = aMale * xMale + bMale
xFemale = np.array([1,5])
yFemale = aFemale * xFemale + bFemale

#linear regression line is added to plot
plt.plot(xMale, yMale, color = "blue")
plt.plot(xFemale, yFemale, color = "orange")
                  
#title is added
plt.title("Heart Weight vs. Body Weight for Male and Female Cats", fontsize = 24)

#x and y label are created
plt.xlabel("Body Weight (kg)", fontsize = 14)
plt.ylabel("Heart Weight (g)", fontsize = 14)

#legend is created
plt.legend()

#print statements with the slope
print("With every additional kilo of body weight, males would add more heart mass.")
print("With every additional kilo of body weight, males would add", aMale, 'g more heart mass')
print("\n")

#show the plot
plt.show()
```

## (E) Calculate a body weight using a heart weight

You have just found an record of an old **female** cat heart specimen weighting 15 g. Can you use your linear fit model to predict the cat's weight?  Provide informative printout.

```python
#we are looking for a female cat heart specimen weighing 15g
yFemale = 15

#formula to solve for x
xFemale = (yFemale-bFemale) / aFemale

#x is printed
print("Using our linear fit model, we predict a female cat heart weighing", yFemale, "g to weigh ", xFemale, "kg")
```
