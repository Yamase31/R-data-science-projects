### (A) What is the GC content of dnaX?

In molecular biology and genetics, GC-content (or guanine-cytosine content) is the percentage of nitrogenous bases on a DNA or RNA molecule that are either guanine(G) or cytosine(C). GC-content is often used as a measure of stability of specific DNA. Below is an example of the calculation of GC content for short DNA fragment

*Write a code that prints out the GC content of this DNA sequence (that is the proportion of bases that are either G or C). *

```python
gCount = dnaX.count("G")
cCount = dnaX.count("C")
dnaXLength = len(dnaX)
GCTotal = gCount + cCount
GCContent = GCTotal / dnaXLength
#conver to percentage
GCContent = GCContent * 100
print("The GC content of this DNA is " + str(GCContent) + "%")
```

### (B) What is the complementary sequence for dnaX?

 *Write a code that prints out the complementary sequence for dnaX. That is, you need to replace each base with its pair: **A** with **T**, **T** with **A**, **C** with **G**, and **G** with **C***. Hint: think carefully about how you are going to do this. It is a little tricky.

```python
print("Original Sequence: " + dnaX)
compSeq = str(dnaX)
compSeq = compSeq.replace("A", "1")
compSeq = compSeq.replace("T", "2")
compSeq = compSeq.replace("C", "3")
compSeq = compSeq.replace("G", "4")

compSeq = compSeq.replace("1", "T")
compSeq = compSeq.replace("2", "A")
compSeq = compSeq.replace("3", "G")
compSeq = compSeq.replace("4", "C")
print("Complementart Sequence: " + compSeq)
```

###  (C) Does dnaX contain a TATA box?

*The sequence "TATA" (also known as TATA box) is very important for molecular biologists. It indicates the binding site for RNA transcription. Does dnaX contains a CCAT box? If yes, in which position is it located? *
 
**Note**: Be careful with your numbering system here: python uses index 0 for the first element of the sequence, while molecular biologists start counting from position 1.

```python
TATALocation = dnaX.find("TATA")
#converts to biology counting standards
TATALocationStart = TATALocation + 1
TATALocationEnd = TATALocationStart + 3
print("The TATA box is located between position " + str(TATALocationStart) + " and " + str(TATALocationEnd))
```

### (D)  What is the coding region of dnaX?

Introns and exons are nucleotide sequences within a gene. Introns are removed by RNA splicing as RNA matures, meaning that they are not expressed in the final protein product, while exons go on to became a protein. Therefore, only sequencies encoding an exon are referred to as **coding sequences**.

Suppose that ***dnaX*** *contains two exons and one intron. The first exon runs from the start of the sequence to 11-th element, and the second exon  runs from the 24-th element to the end of the sequence. Write a code that prints just the coding region of ***dnaX***

```python
dnaX1 = dnaX[:11]
dnaX2 = dnaX[23:]
dnaExons = dnaX1 + dnaX2
print("Coding region of dnaX is " + dnaExons)
```

### (E) What percentage of dnaX contains coding sequence?

*Using the data from previous parts, write a code that will calculate what percentage of **dnaX** sequence is coding. Remember, I dont' want you to just do some math with numbers. You should be using variables names and calculations using those variables.* 

```python
dnaExonsLength = len(dnaExons)
codingPercentage = dnaExonsLength / dnaXLength
codingPercentage = codingPercentage * 100
print("The percentage of dnaX that is a coding sequence is " + str(codingPercentage) + "%")
```
